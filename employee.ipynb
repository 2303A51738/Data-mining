```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
file_path = 'employee_data.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Display the dataset before preprocessing
print("Dataset before preprocessing:")
print(df.head())
```

```python
# Mean imputation for 'Age'
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Median imputation for 'Salary'
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

# Mode imputation for 'Job_Satisfaction'
df['Job_Satisfaction'] = df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mode()[0])

print("\nDataset after handling missing values:")
print(df.head())
```

```python
# Min-Max Scaling for 'Age'
df['Age_Scaled'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())

# Min-Max Scaling for 'Salary'
df['Salary_Scaled'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())

print("\nDataset after Min-Max Scaling:")
print(df[['Age', 'Age_Scaled', 'Salary', 'Salary_Scaled']].head())
```

```python
# Min-Max Scaling for 'Age'
df['Age_Scaled'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())

# Min-Max Scaling for 'Salary'
df['Salary_Scaled'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())

print("\nDataset after Min-Max Scaling:")
print(df[['Age', 'Age_Scaled', 'Salary', 'Salary_Scaled']].head())
```

```python
# Display final preprocessed dataset
print("\nFinal Preprocessed Dataset:")
print(df.head())
```

```python
# Save the preprocessed dataset
df.to_csv('preprocessed_employee_data.csv', index=False)

print("\nPreprocessed dataset saved as 'preprocessed_employee_data.csv'")
```
